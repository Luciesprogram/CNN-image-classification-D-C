```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
import numpy as np
```

## Load Dataset

```python
import pandas as pd


X_train = pd.read_csv('/content/drive/MyDrive/input.csv')
y_train = pd.read_csv('/content/drive/MyDrive/labels.csv')

X_test = pd.read_csv('/content/drive/MyDrive/input_test.csv')
y_test = pd.read_csv('/content/drive/MyDrive/labels_test.csv')
```

```python
print("X_train: ",X_train.shape)
print("X_test: ",y_train.shape)
print("y_train: ",X_test.shape)
print("y_test: ",y_test.shape)
```

```python
X_train = X_train.values.reshape(len(X_train), 100, 100, 3)
y_train = y_train.values.reshape(len(y_train), 1)

X_test = X_test.values.reshape(len(X_test), 100, 100, 3)
y_test = y_test.values.reshape(len(y_test),1)

```

```python
X_train = X_train/255.0
X_test = X_test/255.0
```

```python
print("X_train: ",X_train.shape)
print("X_test: ",y_train.shape)
print("y_train: ",X_test.shape)
print("y_test: ",y_test.shape)
```

```python
import matplotlib.pyplot as plt
import random

idx = random.randint(0,len(X_train))
plt.imshow(X_train[idx,:])
plt.axis('off')
plt.show
```

# **Model**

```python
 model = Sequential([
     Conv2D(32, (3,3),activation='relu',input_shape=(100,100,3)),
     MaxPooling2D((2,2)),

     Conv2D(32,(3,3), activation='relu'),
     MaxPooling2D((2,2)),

     Flatten(),
     Dense(64, activation='relu'),
     Dense(1,activation='sigmoid')
 ])
```

```python
model = Sequential()

model.add(Conv2D(32,(3,3),activation='relu',input_shape=(100,100,3)))
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())
model.add(Dense(64,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
import tensorflow as tf
opt = tf.keras.optimizers.SGD(learning_rate=0.001)
model.compile(loss = 'binary_crossentropy',optimizer=opt,metrics = ['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10,batch_size=64)
```

```python
model.evaluate(X_test,y_test)
```

## Making Predictions

```python
idx2 = random.randint(0,len(y_test))
plt.imshow(X_test[idx2,:])
plt.show()

y_pred = model.predict(X_test[idx2, :].reshape(1, 100, 100, 3))
y_pred = y_pred > 0.5

if (y_pred == 0):
  pred = 'Dog'
else:
  pred ='Cat'
print("Model Predicting : ",pred)
```
